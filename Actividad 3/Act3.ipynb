```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car_accidents.csv")  #Cargar los datos...
df
```

```python
df.hist(column="Humidity(%)", grid=False, bins=7, orientation="vertical", color = "b")
plt.show()
```

```python
df.hist(column="Temperature(F)", grid=False, bins=7, orientation="vertical", color = "r")
plt.show()
```

```python
df.boxplot(column="Humidity(%)", color = "green", showmeans=True )
plt.show()
```

```python
# Con outliers
df.boxplot(column="Temperature(F)", color = "green", showmeans=True )
plt.show()
```

```python
df1 = df[df["Temperature(F)"] < 61] #Quitando los outliers...
df1.boxplot(column="Temperature(F)", color = "green", showmeans=True )
plt.show()
```

```python
# Heatmap de correlaciones entre todas las variables numÃ©ricas
plt.figure(figsize=(15, 5))
sns.heatmap(df1.corr(), annot=True, vmin=0, vmax=1, cmap="cividis");
plt.show()
```
