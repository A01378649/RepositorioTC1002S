```python
c1 = df["Severity"]     #Para los valores de "Severidad"
print(c1.unique())      #Valores únicos
print(c1.max())         #Valor máximo
print(c1.min())         #Valor mínimo
print(c1.mean())        #Media
print(c1.median())      #Mediana
print(c1.std())         #Desviación estandar
```
